```python
import psycopg2
import pandas as pd
```

```python
def create_database():
    # connect to default database
    #conn = psycopg2.connect("host=localhost dbname=studentdb user=student password=student")
    conn = psycopg2.connect("host=localhost dbname=postgres user=postgres password=root")
    conn.set_session(autocommit=True)
    cur = conn.cursor()
    
    # create sparkify database with UTF8 encoding
    cur.execute("DROP DATABASE IF EXISTS schoolrecords")
    cur.execute("CREATE DATABASE schoolrecords")
    
    # close connection to default database
    conn.close
    
    #connect to sparkify database
    conn = psycopg2.connect("host=localhost dbname=schoolrecords user=postgres password=root")
    cur = conn.cursor()
    
    return cur, conn
```
